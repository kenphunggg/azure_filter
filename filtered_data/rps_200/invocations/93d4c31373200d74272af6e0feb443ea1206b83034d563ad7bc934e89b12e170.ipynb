```python
import os
import csv
import numpy as np
import matplotlib.pyplot as plt

# Define the filename you want to open
filename = 'invocations_per_function_md.anon.d01.csv'
current_folder = os.getcwd()
full_path = os.path.join(current_folder, 'filtered_data', 'rps_200', 'invocations', filename)

# Open and read the file using the correct, full path

for i in range(14):
    filename = f'invocations_per_function_md.anon.d{i+1:02d}.csv'
    current_folder = os.getcwd()
    full_path = os.path.join(current_folder, 'filtered_data', 'rps_200', 'invocations', filename)

    with open(full_path) as file:
        reader = csv.reader(file, delimiter=',')

        for row in reader:
            if row[0] == 'ffdd3d1c6a84caf0b741293bfac9d0d9c6445b53dfc29417d196981d0bb1192a' and row[1] == '228ef3c9f06cde19b4382e39c9339cc9b62b521ae71d70ed02d775df20529a43' and row[2] == '93d4c31373200d74272af6e0feb443ea1206b83034d563ad7bc934e89b12e170':
                y_point_arr = row[4:]
                
                # Convert data to a numeric type for plotting
                x_points = np.arange(0, 1440)
                y_points = np.array(y_point_arr, dtype=np.float64)
                
                pattern_x_points = np.linspace(0, 1440, 1440)
                pattern_y_points = np.polyval(np.polyfit(x_points, y_points, 3), pattern_x_points)
                
                plt.plot(x_points, y_points, label='data point')
                plt.plot(pattern_x_points, pattern_y_points, label='patern')
                plt.title(f'Day {i+1}')
                plt.show()
```
